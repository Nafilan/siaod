```python
import sys

def maxSubArray(a):
    size = len(a)
    max_so_far = -sys.maxsize - 1
    max_ending_here = 0
    start = 0
    end = 0
    s = 0
    for i in range(0, size):
        max_ending_here += a[i]
        if max_so_far < max_ending_here:
            max_so_far = max_ending_here
            start = s
            end = i
        if max_ending_here < 0:
            max_ending_here = 0
            s = i+1
    if max_so_far > 0:
        return start, end
    else:
        return 0, -1

def task3(arr):
    shift = []
    for i in range(1, len(arr), 2):
        shift.append(arr[i] - arr[i - 1])
    start, end = maxSubArray(shift)
    start *= 2
    end = end * 2 + 1
    l = arr[:start]
    mid = arr[start:end + 1]
    mid.reverse()
    r = arr[end + 1:]
    l.extend(mid)
    l.extend(r)
    sum1 = 0
    for i in range(0, len(l), 2):
        sum1 += l[i]
    
    shift = []
    for i in range(2, len(arr), 2):
        shift.append(arr[i - 1] - arr[i])
    start, end = maxSubArray(shift)
    start = start * 2 + 1
    end = end * 2 + 1 + 1
    l = arr[:start]
    mid = arr[start:end + 1]
    mid.reverse()
    r = arr[end + 1:]
    l.extend(mid)
    l.extend(r)
    sum2 = 0
    for i in range(0, len(l), 2):
        sum2 += l[i]
    return max(sum1, sum2)

print("–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö:")
res = []
t = int(input("t = "))
if t not in range(1, 2 * 10 ** 4 + 1):
    sys.exit("t –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=t<=2*10^4")
nsum = 0
for i in range(t):
    n = int(input("n = "))
    nsum += n
    if nsum > 2 * 10 ** 5:
        sys.exit("–°—É–º–º–∞ n –ø–æ –≤—Å–µ–º –Ω–∞–±–æ—Ä–∞–º –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 2*10^5")
    if n not in range(1, 2 * 10 ** 5 + 1):
        sys.exit("n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=n<=2*10^5")
    arr = input("–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤ ").split()
    arr = [int(arr[i]) for i in range(len(arr))]
    if len(arr) != n:
        sys.exit("–î–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞ –Ω–µ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å n")
    for item in arr:
        if item not in range(1, 10 ** 9):
            sys.exit("–≠–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=ai<=10^9")
    res.append(task3(arr))
print("–í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:")
for i in range(t):
    print(res[i])
```

```python
import sys
import math
import itertools

def task5(arr, k):
    alice_bob = []
    alice = []
    bob = []
    for item in arr:
        if item[1] and item[2]:
            alice_bob.append(item[0])
        elif item[1] and not item[2]:
            alice.append(item[0])
        elif not item[1] and item[2]:
            bob.append(item[0])
    alice_bob.sort()
    alice.sort()
    bob.sort()
    minsum = math.inf 
    alice_bob.insert(0, 0)
    alice.insert(0, 0)
    bob.insert(0, 0)
    alice_bob = list(itertools.accumulate(alice_bob))
    alice = list(itertools.accumulate(alice))
    bob = list(itertools.accumulate(bob))
    for i in range(k + 1):
        if i not in range(len(alice_bob)):
            break
        books_left = k - i
        if books_left not in range(len(alice)) or books_left not in range(len(bob)):
            continue
        
        if alice_bob[i] + alice[books_left] + bob[books_left] < minsum:
            minsum = alice_bob[i] + alice[books_left] + bob[books_left]

    if minsum == math.inf:
        return -1
    return minsum

print("–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö:")
n, k = input("n, k = ").split()
n, k = int(n), int(k)
if not 1 <= k <= n <= 2 * 10 ** 5:
    sys.exit("–ù–µ—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —É—Å–ª–æ–≤–∏—é 1<=k<=n<=2*10^5")
arr = []
for i in range(int(n)):
    t, a, b = input("t, a, b = ").split()
    t, a, b = int(t), int(a), int(b)
    if t not in range(1, 10 ** 4 + 1):
        sys.exit("t –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=t<=10^4")
    if a != 0 and a != 1:
        sys.exit("a –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 0<=a<=1")
    if b != 0 and b != 1:
        sys.exit("b –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 0<=b<=1")
    arr.append([t, a, b])
print("–í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:")
print(task5(arr, k))
```

```python
import sys
import math
import itertools

def task6(arr, a, b):
    arr.insert(0, 0)
    arr.insert(0, 0)
    mincost = math.inf
    arr_acc = list(itertools.accumulate(arr))
    for i in range(1, len(arr) - 1):
        cost = a * arr[i] + b * arr_acc[-1] - b * (arr_acc[i - 1] + arr[i] * (len(arr) - i - 1))
        if cost < mincost:
            mincost = cost
    return mincost

print("–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö:")
result = []
t = int(input("t = "))
if t not in range(1, 1001):
    sys.exit("t –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=t<=1000")
nsum = 0
for i in range(t):
    n, a, b = input("n, a, b = ").split()
    n, a, b = int(n), int(a), int(b)
    nsum += n
    if nsum > 2 * 10 ** 5:
        sys.exit("–°—É–º–º–∞ n –ø–æ –≤—Å–µ–º –Ω–∞–±–æ—Ä–∞–º –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 2*10^5")
    if n not in range(1, 2 * 10 ** 5 + 1):
        sys.exit("n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=n<=2*10^5")
    if a not in range(1, 10 ** 5 + 1):
        sys.exit("a –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=a<=10^5")
    if b not in range(1, 10 ** 5 + 1):
        sys.exit("b –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=b<=10^5")
    arr = input("–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤ ").split()
    arr = [int(arr[i]) for i in range(len(arr))]
    if len(arr) != n:
        sys.exit("–î–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞ –Ω–µ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å n")
    prev = 0
    for item in arr:
        if item <= prev or item not in range(1, 10 ** 8 + 1):
            sys.exit("–ù–µ—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —É—Å–ª–æ–≤–∏—é 1<=x1")
        prev = item
    result.append(task6(arr, a, b))
print("–í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:")
for item in result:
    print(item)
```

–í —Ç—Ä–µ—Ç—å–µ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ ùëö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ùëè1,‚Ä¶,ùëèùëö (1‚â§ùëèùëó‚â§1018).

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í—ã–≤–µ–¥–∏—Ç–µ ùëö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª. ùëó-–µ –∏–∑ —ç—Ç–∏—Ö —á–∏—Å–µ–ª –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ä–∞–≤–Ω–æ –ù–û–î(ùëé1+ùëèùëó,‚Ä¶,ùëéùëõ+ùëèùëó).

```python
import sys

def gcd(a, b):
    while a != 0 and b != 0:
        if a > b:
            a = a % b
        else:
            b = b % a
    return a + b

def gcd_arr(arr):
    result = arr[0]
    for i in range(1, len(arr)):
        result = gcd(result, arr[i])
    return result

def task7(a, b):
    a = list(map(int, a))
    a.sort()
    b = list(map(int, b))
    arr = [a[i] - a[0] for i in range(0, len(a))]
    gcd_part = gcd_arr(arr)
    result = []
    for j in range(len(b)):
        result.append(gcd(a[0] + b[j], gcd_part))
    return result

print("–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö:")
n, m = input("n, m = ").split()
n, m = int(n), int(m)
if n not in range(1, 2 * 10 ** 5 + 1):
    sys.exit("n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=n<=2*10^5")
if m not in range(1, 2 * 10 ** 5 + 1):
    sys.exit("m –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=m<=2*10^5")
a = input("a = ").split()
a = [int(a[i]) for i in range(len(a))]
if len(a) != n:
    sys.exit("–î–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞ a –Ω–µ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å n")
for item in a:
    if item not in range(1, 10 ** 18 + 1):
        sys.exit("ai –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=ai<=10^18")
b = input("b = ").split()
b = [int(b[i]) for i in range(len(b))]
if len(b) != m:
    sys.exit("–î–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞ b –Ω–µ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å m")
for item in b:
    if item not in range(1, 10 ** 18):
        sys.exit("bj –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=bj<=10^18")
result = task7(a, b)
print("–í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:")
for item in result:
    print(item, end = ' ')
```

```python
import math
import heapq
import sys

def task10(w1):
    w1 = [-item for item in w1]
    w2 = [sum(w1)]
    heapq.heapify(w1)
    heapq.heapify(w2)
    while len(w1) != 0 and len(w2) != 0:
        if w1[0] < w2[0]:
            return "NO"
        elif w1[0] == w2[0]:
            heapq.heappop(w1)
            heapq.heappop(w2)
        else:
            item = heapq.heappop(w2)
            heapq.heappush(w2, math.ceil(item / 2))
            heapq.heappush(w2, math.floor(item / 2))
    return "YES"

print("–í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö:")
result = []
t = int(input("t = "))
if t not in range(1, 10 ** 4):
    sys.exit("t –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=t<=10^4")
nsum = 0
for i in range(t):
    n = int(input("n = "))
    nsum += n
    if nsum > 2 * 10 ** 5:
        sys.exit("–°—É–º–º–∞ n –ø–æ –≤—Å–µ–º –Ω–∞–±–æ—Ä–∞–º –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 2*10^5")
    if n not in range(1, 2 * 10 ** 5 + 1):
        sys.exit("n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=n<=2*10^5")
    arr = input("–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤ ").split()
    arr = [int(arr[i]) for i in range(len(arr))]
    if len(arr) != n:
        sys.exit("–î–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞ –Ω–µ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å n")
    for item in arr:
        if item not in range(1, 10 ** 9):
            sys.exit("ai –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 1<=ai<=10 ** 9")
    result.append(task10(arr))
print("–í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:")
for item in result:
    print(item)
```
