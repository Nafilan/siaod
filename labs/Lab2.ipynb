```python
#Интерполяционный
text = copy.deepcopy(input)

def search(input, required):
    input.sort()
    l = 0
    r = len(input) - 1
    while input[l] < required and input[r] > required:
        if input[l] == input[r]:
            break
        i = (((required - input[l]) * (l - r)) // (input[l] - input[r])) + l
        if input[i] == required:
            return i
        if input[i] > required:
            r = i - 1
            continue
        if input[i] < required:
            l = i + 1
    if input[l] == required:
        return l
    if input[r] == required:
        return r
    return None

def search_delete(input, required):
    i = search(input, required)
    if i:
        input.pop(i)
    return input

def search_append(input, required):
    input.sort()
    l = 0
    r = len(input) - 1
    while input[l] < required and input[r] > required:
        if input[l] == input[r]:
            break
        i = (((required - input[l]) * (l - r)) // (input[l] - input[r])) + l
        if input[i] == required:
            input.insert(i, required)
            return input
        if input[i] > required:
            r = i - 1
            continue
        if input[i] < required:
            l = i + 1
    if input[l] > required:
        input.insert(l, required)
    elif input[r] < required:
        input.insert(r + 1, required)
    return input

t1 = time.time()
print(search(text, 10))
t2 = time.time()
print(search_delete(text, 10))
print(search_append(text, 25))
search_time = round((t2 - t1) * 1000000000)
print("--- {0} ns ---".format(search_time))
```
