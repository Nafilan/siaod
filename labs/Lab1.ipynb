```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
def row_sort(line):
    for i in range(len(line)):
        temp = line[i]
        index = i
        for j in range(i, len(line)):
            if line[j] < temp:
                temp = line[j]
                index = j
        line[index], line[i] = line[i], temp


matrix = copy.deepcopy(random_matrix)                

for line in matrix:
    line = row_sort(line)

print_mat(matrix)
row_sort_time = (round((time.time() - start_time)*1000))
print("--- {0} ms ---".format(row_sort_time))
```

```python
# Сортировка вставкой.
start_time = time.time()
def insert_sort(line):
    for i in range(1, len(line)):
        for j in range(i, 0, -1):
            if line[j] < line[j - 1]:
                line[j], line[j - 1] = line [j - 1], line[j]
            else:
                break

matrix = copy.deepcopy(random_matrix)

for line in matrix:
    line = insert_sort(line)
    
print_mat(matrix)
insert_sort_time = (round((time.time() - start_time)*1000))
print("--- {0} ms ---".format(insert_sort_time))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
def bubble_sort(line):
    n = 1
    for i in range (len(line) - 1):
        for j in range(len(line) - n):
            if line[j] > line[j + 1]:
                line[j], line[j + 1] = line[j + 1], line[j]
        n += 1

matrix = copy.deepcopy(random_matrix)

for line in matrix:
    line = bubble_sort(line)
    
print_mat(matrix)
bubble_sort_time = (round((time.time() - start_time)*1000))
print("--- {0} ms ---".format(bubble_sort_time))
```

```python
# Сортировка Шелла.
start_time = time.time()
def shell_sort(line):
    step = len(line) // 2
    while step > 0:
        for i in range(step, len(line)):
            location = i
            value = line[i]
            while location >= step and line[location - step] > value:
                line[location] = line[location - step]
                location -= step
                line[location] = value
        step = step // 2
        
matrix = copy.deepcopy(random_matrix)

for line in matrix:
    line = shell_sort(line)
    
print_mat(matrix)
shell_sort_time = (round((time.time() - start_time)*1000))
print("--- {0} ms ---".format(shell_sort_time))
```

```python
# Быстрая сортировка.
start_time = time.time()

def quick_sort(line):
    less = []
    equal = []
    greater = []
    
    if len(line) > 1:
        help_line = line[0]
        for num in line:
            if num < help_line:
                less.append(num)
            elif num == help_line:
                equal.append(num)
            elif num > help_line:
                greater.append(num)
        return quick_sort(less)+equal+quick_sort(greater) 
    else:  
        return line
        
matrix = copy.deepcopy(random_matrix)

new_matrix = []
for line in matrix:
    new_matrix.append(quick_sort(line))

print_mat(new_matrix)
quick_sort_time = (round((time.time() - start_time)*1000))
print("--- {0} ms ---".format(quick_sort_time))  
```

```python
# Турнирная сортировка.
start_time = time.time()

def heapify(line,n,index):
    greatest = index
    left = 2*index + 1
    right = 2*index + 2 
    if left < n and line[index] < line[left]:
        greatest = left

    if right < n and line[greatest] < line[right]:
        greatest = right   
        
    if greatest != index:
        line[index],line[greatest] = line[greatest],line[index]  
        heapify(line, n, greatest)

def heap_sort(line):
    n = len(line)
    for i in range(n // 2 - 1, -1, -1):
        heapify(line, n, i)
        
    for i in range(n-1, 0, -1):
        line[i], line[0] = line[0], line[i] 
        heapify(line, i, 0)
    
matrix = copy.deepcopy(random_matrix)
for line in matrix:
      line = heap_sort(line)

print_mat(matrix)  
heap_sort_time = (round((time.time() - start_time)*1000))
print("--- {0} ms ---".format(heap_sort_time))
```
